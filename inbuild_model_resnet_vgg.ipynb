# Classify ImageNet classes with ResNet50

```python
def classify(model):
    main_path = %pwd
    img_path = r'C:\Users\rajeevranjan.sharma\project\dl\data/'
    total_list = os.listdir(img_path)
    len(total_list)
    ext = ['jpg', 'png', 'tif']

    for enty in total_list:
        if enty.split('.')[1] in ext:
            img = image.load_img(img_path+enty, target_size=(224,224))
            np_x = image.img_to_array(img)
            x = np.expand_dims(np_x, axis=0)
            x = preprocess_input(x)

            pred = model.predict(x)

            print("img prediction: ", decode_predictions(pred, top=3)[0])
            #show images
            plt.imshow(img, cmap='gray')
            plt.show()
            
```

# # Build VGG over a custom input tensor

```python
def classify(model):
    main_path = %pwd
    img_path = r'C:\Users\rajeevranjan.sharma\project\dl\data/'
    total_list = os.listdir(img_path)
    len(total_list)
    ext = ['jpg', 'png', 'tif']

    for enty in total_list:
        if enty.split('.')[1] in ext:
            img = image.load_img(img_path+enty, target_size=(300,300))
            np_x = image.img_to_array(img)
            x = np.expand_dims(np_x, axis=0)
            x = preprocess_input(x)

            pred = model.predict(x)
            
            print("img prediction: ", np.argmax(pred))
            #print("img prediction: ", decode_predictions(pred)[0][0])
            #show images
            plt.imshow(img, cmap='gray')
            plt.show()
    return x
```

```python
model = VGG16(include_top=False, input_shape=(300, 300, 3))#weights='imagenet', include_top=False)
model.summary()
```

```python
from keras.layers import Dense
from keras.layers import Flatten

flat1 = Flatten()(model.output)
class1 = Dense(1024, activation='relu')(flat1)
output = Dense(10, activation='softmax')(class1)

#define new model
model = Model(inputs=model.inputs, outputs=output)
model.summary()
```

```python
classify(model)
```
